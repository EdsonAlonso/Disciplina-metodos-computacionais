## Complex exponential

Let's first define the [Euler's formula](https://en.wikipedia.org/wiki/Euler%27s_formula):

<a id='eq17a'></a>

$$
e^{\, i \, \theta} = \cos \left( \theta \right) + i \, \sin\left( \theta \right) \: , \tag{17a}
$$

which can be graphically represented as follows:

```python
mfun.euler_formula(0.37*np.pi)
```

**<a id='fig6'>Figure 6</a>**: *Schematical representation of the Euler's formula. The thick black line represents a unit radius with a horizontal component $\cos \left( \theta \right)$ (red line) and a vertical component $\sin \left( \theta \right)$ (blue line). The angle $\theta$ is measured from the red line to the black line and is represented by the curved black line.*

Consider the Euler's formula ([equation 17a](#eq17a)) for an angle $-\theta$. In this case, according to equations [1](https://nbviewer.jupyter.org/github/birocoles/Disciplina-metodos-computacionais/blob/master/Content/fourier_1D_intro1.ipynb#eq1) and [2](https://nbviewer.jupyter.org/github/birocoles/Disciplina-metodos-computacionais/blob/master/Content/fourier_1D_intro1.ipynb#eq2) of the part 1, we obtain:

<a id='eq17b'></a>

$$
\begin{split}
e^{\, i \, \left( -\theta \right)} &= \overbrace{\cos \left( -\theta \right)}^{\text{even function}} + 
i \, \overbrace{\sin\left( -\theta \right)}^{\text{odd function}} \\
&= \cos \left( \theta \right) - i \, \sin\left( \theta \right)
\end{split} \: . \tag{17b}
$$

By adding equations [17a](#eq17a) and [17b](#eq17b), we obtain 

<a id='eq18a'></a>

$$
\cos \left( \theta \right) = \frac{1}{2} \left( e^{\, i \, \theta} + e^{\, -i \, \theta}\right) \: , \tag{18a}
$$

By subtracting equation [17b](#eq17b) from [17a](#eq17a), we obtain 

<a id='eq18b'></a>

$$
\sin \left( \theta \right) = \frac{1}{2 \, i} \left( e^{\, i \, \theta} - e^{\, -i \, \theta}\right) \: . \tag{18b}
$$

## Fourier series defined as a combination of complex exponentials for describing functions with period $2 \pi$

By substituting equations [18a](#eq18a) and [18b](#eq18b) into [equation 6](https://nbviewer.jupyter.org/github/birocoles/Disciplina-metodos-computacionais/blob/master/Content/fourier_1D_1.ipynb#eq6) of the part 1 and rearranging the terms, we can rewrite the Fourier series as a function of complex exponentials as follows:

<a id='eq19a'></a>

$$
f \left( x \right) = c_{0} + \
\sum \limits_{n=1}^{N} c_{n} e^{\, i \, n \, x} + \
\sum \limits_{n=1}^{N} c_{n}^{\ast} e^{\, -i \, n \, x} \
\: , \quad N \to \infty \: , \tag{19a}
$$

where

<a id='eq19b'></a>

$$
c_{n} = \frac{1}{2} \left( a_{n} - i \, b_{n} \right) \: , \tag{19b}
$$

<a id='eq19c'></a>

$$
c_{n}^{\ast} = \frac{1}{2} \left( a_{n} + i \, b_{n} \right) \: , \tag{19c}
$$

and

<a id='eq19d'></a>

$$
c_{0} = \frac{a_{0}}{2} \: , \tag{19d}
$$

which is obtained by substituting $n = 0$ in equations [19b](#eq19b) and [19c](#eq19c). The coefficients $c_{n}$ ([equation 19b](#eq19b)) and $c_{n}^{\ast}$ ([equation 19c](#eq19c)) are complex functions of the integer $n$. In [equation 19a](#eq19a), these complex functions are evaluated with $n > 0$. However, they can also be evaluated with $n < 0$.

Let's analyse the particular case in which the coefficient $c_{n}$ ([equation 19b](#eq19b)) and, consequently, the coefficients $a_{n}$ ([equation 15](https://nbviewer.jupyter.org/github/birocoles/Disciplina-metodos-computacionais/blob/master/Content/fourier_1D_1.ipynb#eq15)) and $b_{n}$ ([equation 16](https://nbviewer.jupyter.org/github/birocoles/Disciplina-metodos-computacionais/blob/master/Content/fourier_1D_1.ipynb#eq16)) are calculated by using $n < 0$. 

The coefficients $a_{n}$ ([equation 15](https://nbviewer.jupyter.org/github/birocoles/Disciplina-metodos-computacionais/blob/master/Content/fourier_1D_1.ipynb#eq15)) do not depend on the signal of $n$ because they are defined by a cosine, which is an even function. On the other hand, the coefficients $b_{n}$ ([equation 16](https://nbviewer.jupyter.org/github/birocoles/Disciplina-metodos-computacionais/blob/master/Content/fourier_1D_1.ipynb#eq16)) depends on the signal of $n$ because they are defined by a sine, which is an odd function. Consequently,

<a id='eq20c'></a>

$$
c_{n} \:\: \text{for} \:\: n < 0 \: = \: c_{n}^{\ast} \:\: \text{for} \:\: n > 0 \quad . \tag{20c}
$$

This equation shows that:

$$
c_{n} \, e^{\, i \, n \, x} \: , \quad \text{for} \:\: n < 0 \: ,
$$

is equal to 

$$
c_{n}^{\ast} \, e^{\, -i \, n \, x} \: , \quad \text{for} \:\: n > 0 \: .
$$

As a consequence, the second summation in [equation 19a](#eq19a) can be rewritten as follows:

$$
\sum \limits_{n=1}^{N} c_{n}^{\ast} \, e^{\, -i \, n \, x} \ =
\sum \limits_{n=-1}^{-N} c_{n} \, e^{\, i \, n \, x} \quad ,
$$

and [equation 19a](#eq19a) can be rewritten in an alternative form given by

<a id='eq20d'></a>

$$
f(x) = \sum \limits_{n=-N}^{N} c_{n} e^{\, i \, n \, x} \: , \quad \text{for} \; N \to \infty \: . \tag{20d}
$$

<a id='subsec4.1'></a>

### Calculating $c_{n}$ by using the orthogonality relations between complex exponentials

Let $m$ and $n$ be integers, e. g., $m \, , \, n \in \mathbb{N}$. The orthogonality relations between exponential functions are given by:

<a id='eq21a'></a>

$$
\int \limits_{-\pi}^{\pi} e^{\, -i \, m \, x} \, e^{\, i \, n \, x} \, dx = \text{A} + \text{B} + \text{C} \: , \tag{21a}
$$

where, according to equations [17a](#eq17a) and [17b](#eq17b),

<a id='eq21b'></a>

$$
\text{A} = \int \limits_{-\pi}^{\pi} \cos \left( m \, x \right) \, \cos \left( n \, x \right) \; dx \: , \tag{21b}
$$

<a id='eq21c'></a>

$$
\text{B} = i \int \limits_{-\pi}^{\pi} \cos \left( m \, x \right) \, \sin \left( n \, x \right) \; dx - \
i \int \limits_{-\pi}^{\pi} \sin \left( m \, x \right) \, \cos \left( n \, x \right) \; dx \: , \tag{21c}
$$

and

<a id='eq21d'></a>

$$
\text{C} = \int \limits_{-\pi}^{\pi} \sin \left( m \, x \right) \, \sin \left( n \, x \right) \; dx \: . \tag{21d}
$$

For $m \ne n$, $\text{A} = 0$ ([equation 21b](#eq21b)), $\text{B} = 0$ ([equation 21c](#eq21c)), $\text{C} = 0$ ([equation 21d](#eq21d)), and, consequently, 

<a id='eq22a'></a>

$$
\int \limits_{-\pi}^{\pi} e^{\, i \, -m \, x} \, e^{\, i \, n \, x} \, dx = 0 \: . \tag{22a}
$$

For $m = n \ne 0$, $\text{A} = \pi$ ([equation 21b](#eq21b)), $\text{B} = 0$ ([equation 21c](#eq21c)), $\text{C} = \pi$ ([equation 21d](#eq21d)), and, consequently, 

<a id='eq22b'></a>

$$
\int \limits_{-\pi}^{\pi} e^{\, -i \, m \, x} \, e^{\, i \, n \, x} \, dx = 2 \, \pi \: . \tag{22b}
$$

For $m = n = 0$, $\text{A} = 2 \, \pi$ ([equation 21b](#eq21b)), $\text{B} = 0$ ([equation 21c](#eq21c)), $\text{C} = 0$ ([equation 21d](#eq21d)), and, consequently, 

<a id='eq22c'></a>

$$
\int \limits_{-\pi}^{\pi} e^{\, -i \, m \, x} \, e^{\, i \, n \, x} \, dx = 2 \, \pi \: . \tag{22c}
$$

By multypling [equation 20d](#eq20d) by $e^{\, -i \, m \, x}$ (where $m$ is an integer) and integrating from $-\pi$ to $\pi$, we obtain

$$
\int \limits_{-\pi}^{\pi} e^{\, -i \, m \, x} \, f(x) \; dx \
= \sum \limits_{n=-N}^{N} c_{n} \int \limits_{-\pi}^{\pi} e^{\, -i \, m \, x} \, \
e^{\, i \, n \, x} \; dx \: , \quad \text{for} \; N \to \infty \: ,
$$

which results in

$$
\int \limits_{-\pi}^{\pi} e^{\, -i \, n \, x} \, f(x) \; dx \
= c_{n} \, 2 \, \pi \: ,
$$

and, consequently,

<a id='eq23'></a>

$$
c_{n} = \frac{1}{2 \, \pi} \int \limits_{-\pi}^{\pi} e^{\, -i \, n \, x} \, f(x) \; dx \: . \tag{23}
$$

[Equation 23](#eq23) can be rewritten by using the Euler's formula with negative argument ([equation 17b](#eq17b)):

<a id='eq24'></a>

$$
\begin{split}
c_{n} &= \frac{1}{2 \, \pi} \int \limits_{-\pi}^{\pi} \left[ \cos \left( n \, x \right) - i \sin \left( n \, x \right) \right] \, f(x) \; dx \\
&= \frac{1}{2} \, \left[ \frac{1}{\pi} \int \limits_{-\pi}^{\pi} \cos \left( n \, x \right) \, f(x) \; dx -
i \; \frac{1}{\pi} \int \limits_{-\pi}^{\pi} \sin \left( n \, x \right) \, f(x) \; dx \right] \\
&= \frac{1}{2} \left( a_{n} - i \, b_{n} \right)
\end{split}
\: , \tag{24}
$$

which is in perfect agreement with [equation 19b](#eq19b).
